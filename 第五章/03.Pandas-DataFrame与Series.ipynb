```python
import pandas as pd
#创建DataFrame
df1 = pd.DataFrame([
    {"name": "张三", "Chinese": 18, "Math": 80, "English": 70}
    , {"name": "李四", "Chinese": 19, "Math": 70, "English": 60}
    , {"name": "王五", "Chinese": 20, "Math": 60, "English": 50}
])

df2 = pd.DataFrame({
    "name": ["张三", "李四", "王五"]
    , "Chinese": [18, 19, 20]
    , "Math": [80, 70, 60]
    , "English": [70, 60, 50]
})

df3 = pd.DataFrame(
    [
        ["张三", 18, 80, 70]
        , ["李四", 19, 70, 60]
        , ["王五", 20, 60, 50]
    ]
    , columns=["name", "Chinese", "Math", "English"]
)

df4 = pd.DataFrame([
    ["张三", 18, 80, 70]
    , ["李四", 19, 70, 60]
    , ["王五", 20, 60, 50]
],columns=["name","Chinese","Math","English"],index=[1,2,3])
df4

# DataFrame 常见属性 index columns values size dtype shape
df4.columns.tolist()
df4.values.tolist()
df4.size
df4.dtypes
df4.shape
```
