```python
from matplotlib.axes import Axes
import  matplotlib.pyplot as plt
from pydeck.data_utils.viewport_helpers import bbox_to_zoom_level

#展示中文
plt.rcParams['font.sans-serif'] = ['SimHei']

#创建子图,figure画布,axes是子图对应的数组
figure,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,6))
countries = ['中国','美国','日本','韩国','德国']
values = [20.5,15.2,5.3,4.8,3.9]
axes1: Axes = axes[0]
axes1.set_title('世界石油储备',fontsize=18)
axes1.bar(countries,values,color='lightgreen',width=0.7)
axes1.set_xlabel("国家")
axes1.set_ylabel("石油储备量(亿桶)")
axes1.grid(linestyle='--',alpha=0.3)

#饼状图,比例分析
countries2= ['印度','中国','美国','印尼','巴基斯坦','尼日利亚','巴西','俄罗斯','其他']
values2 = [14.51,14.09,3.4,2.83,2.51,2.33,2.12,1.44,20]
axes2: Axes = axes[1]
axes2.pie(values2,labels=countries2,autopct='%1.1f%%')
axes2.set_title('世界人口比例',fontsize=18)
axes2.legend(loc='lower right',ncol=5,bbox_to_anchor=(1, -0.12))

#保存图片
plt.savefig('data/01.png')
```
